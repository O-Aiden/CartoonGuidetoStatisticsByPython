## 正态分布

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def gd(x, mu=0, std=1):
    """
    x:自变量
    mu:均值
    std: 标准差
    """
    left = 1 / (np.sqrt(2 * math.pi) * std)
    right = np.exp(-(x - mu)**2 / (2 * (std*std)))
    return left * right


if __name__ == '__main__':
    #  自变量
    x = np.arange(0, 100, 0.1)
    #  因变量（不同均值或方差）
    y_1 = gd(x, 53, 15)
    y_2 = gd(x, 53, 5)
    y_3 = gd(x, 30, 5)

    #  绘图
    plt.plot(x, y_1, color='green')
    plt.plot(x, y_2, color='blue')
    plt.plot(x, y_3, color='yellow')
    #  设置坐标系
    plt.xlim(0, 100)
    plt.ylim(0, 0.08)

    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data', 0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data', 0))

    plt.legend(labels=['$\mu = 53, \sigma=15$', '$\mu = 53, \sigma=5$', '$\mu = 30, \sigma=5$'])
    plt.show()
```

## 标准正态分布

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def gd(x, mu=0, std=1):
    """
    x:自变量
    mu:均值
    std: 标准差
    """
    left = 1 / (np.sqrt(2 * math.pi) * std)
    right = np.exp(-(x - mu)**2 / (2 * (std*std)))
    return left * right


if __name__ == '__main__':
    #  自变量
    x = np.arange(-4, 5, 0.1)
    #  因变量（不同均值或方差）
    y_1 = gd(x, 0, 1)

    #  绘图
    plt.plot(x, y_1, color='red')
    #  设置坐标系
    plt.xlim(-5.0, 5.0)
    plt.ylim(-0.2, 1)

    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data', 0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data', 0))

    plt.legend(labels=['$\mu = 0, \sigma=1$'])
    plt.show()
```
