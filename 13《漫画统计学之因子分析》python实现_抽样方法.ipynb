```python
import pandas as pd

data = pd.read_excel('汽车企业销量数据.xlsx',index_col=0)
data.head(5)
```

# 简单随机抽样

# 分层抽样

```python
temp = data[['企业','2021年']]
# 总样本
temp['2021年'].sum()
# 各汽车销量占总样本的比例
temp['频率'] = temp['2021年']/temp['2021年'].sum()
# 设定总过抽样180例
temp['抽样'] = 180*temp['频率']
# 四舍五入
temp['抽样'] = (temp['抽样']+0.5).astype(int)
temp.head(5)
```

# 二阶抽样

```python
temp = data[['企业','2021年']]
# 求累计销量
temp['累计销量'] = temp['2021年'].cumsum()
# 求累计下限
temp['累计销量下限'] = temp['累计销量'] - temp['2021年'] + 1 
# 求累计上限
temp['累计销量上限'] = temp['累计销量']
temp.head(5)
```

## 步骤一：求出10个随机数

```python
import random
 
rands = [random.randint(1,temp['2021年'].sum()) for _ in range(10)]
rands
```

## 步骤二：随机数介于累计上限和下限值之间的汽车企业

```python
names = []
for rand in rands:
  a = temp.loc[(temp['累计销量下限']<rand) & (temp['累计销量上限']>rand),['企业']]
  names.append([int(a.index.values),a['企业'].to_list()[0],int(rand)])
names
```

```python
import random
data.columns = ['企业','五月','四月','三月','二月','一月','上年十二月','2021年', '2020年', '2019年']
temp1 = data[['企业','五月','四月','三月','二月','一月']]
# 将企业列设为索引列
temp1.set_index('企业', inplace=True)
# 删除这五个月总销量为0的企业
temp1 = temp1.loc[(temp1.sum(axis=1) != 0)]
temp1.tail(5)
# 对行列进行转置
temp1_T = temp1.T
temp1_T.head(5)

data_copy = data.copy()
# 删除2021年销量为0的企业
data_copy = data_copy.loc[(data_copy['2021年'] != 0)]
data_copy.set_index('企业', inplace=True)
companys = data_copy.index.values
all_names = []
for company in companys: 
  ## 随机抽取三个月份
  rands = [random.randint(1,data_copy['2021年'][company]) for _ in range(3)]
  # print(rands)

  ## 计算累积上限和下限
  temp = temp1_T[[company]]
  # 求累计销量
  temp['累计销量'] = temp1_T[company].cumsum()
  # 求累计下限
  temp['累计销量下限'] = temp['累计销量'] - temp1_T[company] + 1 
  # 求累计上限
  temp['累计销量上限'] = temp['累计销量']
  
  temp.index.names = ['月份']
  temp = temp.reset_index()
  # print(temp.index)
  
  ## 具体对每个企业的哪几个月销量进行抽样
  names = []
  for rand in rands:
    a = temp.loc[(temp['累计销量下限']<=rand) & (temp['累计销量上限']>=rand),['月份']]
    names.append(a['月份'].to_list()[0])
  all_names.append([company,names])
```

```python
all_names
```

```python
## 对数据进行合并
data_samp = pd.DataFrame(all_names)
data_samp.columns = ['企业','抽样月份']
temp_samp_copy = temp_samp.copy()

temp_samp_copy.set_index('企业', inplace=True)
data_samp.set_index('企业', inplace=True)

result = pd.concat([data_samp,temp_samp_copy], axis=1).dropna()
result['每月抽样'] = (result['抽样']/3 + 0.5).astype(int)
result = result.loc[(result['每月抽样'] != 0)]

result
```
